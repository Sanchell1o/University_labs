### **2. Основные операции с NumPy:**


#### **2.1. Создание векторов и матриц:**
- Создайте векторы и матрицы с использованием np.array(), соответствующие данным из набора. Это могут быть временные ряды или другие важные метрики.

```python
# Временной ряд количества велосипедов
time_series_bike_counts = np.array(dataframe[['DateTime', 'Rented Bike Count']])

#Вектор арендованных велосипедов
bike_count = np.array(dataframe['Rented Bike Count'])

# Вектор температуры
temperature = np.array(dataframe['Temperature(°C)'])

# Вектор влажности
humidity = np.array(dataframe['Humidity(%)'])

# Вектор скорости ветра
wind_speed = np.array(dataframe['Wind speed (m/s)'])

# Вектор видимости
visibility = np.array(dataframe['Visibility (10m)'])

# Вектор количества осадков
rainfall = np.array(dataframe['Rainfall(mm)'])

# Вектор количества снега
snowfall = np.array(dataframe['Snowfall (cm)'])

# Вектор дат
date_vector = dataframe['Date']

# Создание матрицы признаков
feature_matrix = np.array([
    dataframe['Temperature(°C)'],
    dataframe['Humidity(%)'],
    dataframe['Wind speed (m/s)'],
    dataframe['Dew point temperature(°C)']
]).T

# Создание матрицы погодных условий
weather_matrix = np.array([
    dataframe['Rainfall(mm)'],
    dataframe['Snowfall (cm)'],
    dataframe['Solar Radiation (MJ/m2)'].to_numpy()
]).T

# Создание матрицы средних значений (числовая)

numeric_cols = [
    'Rented Bike Count', 'Hour', 'Temperature(°C)', 'Humidity(%)',
    'Wind speed (m/s)', 'Visibility (10m)', 'Dew point temperature(°C)',
    'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)'
]

mean_matrix_numeric = np.array([dataframe[col].mean() for col in numeric_cols])

# Создание матрицы средних значений (бинарная)
binary_cols = [
    'Seasons_Autumn', 'Seasons_Spring', 'Seasons_Summer', 'Seasons_Winter',
    'Holiday_Holiday', 'Holiday_No Holiday', 'Functioning Day_No',
    'Functioning Day_Yes'
]

mean_matrix_binary = np.array([dataframe[col].mean() for col in binary_cols])
```

#### **2.2. Использование матриц и векторов:**
- Используйте матрицы и векторы для представления различных параметров, таких как количество прокачанных велосипедов, температура, влажность и другие переменные.

```python
print("Времянной ряд количества арендованных велосипедов:")
print("Формат:", time_series_bike_counts.shape)
print("Первые 5 элементов:", time_series_bike_counts[:5])
print("-" * 30)

print("Вектор арендованных велосипедов:")
print("Формат:", bike_count.shape)
print("Первые 5 элементов:", bike_count[:5])
print("-" * 30)

print("Вектор температуры:")
print("Формат:", temperature.shape)
print("Первые 5 элементов:", temperature[:5])
print("-" * 30)

print("Вектор влажности:")
print("Формат:", humidity.shape)
print("Первые 5 элементов:", humidity[:5])
print("-" * 30)

print("Вектор скорости ветра:")
print("Формат:", wind_speed.shape)
print("Первые 5 элементов:", wind_speed[:5])
print("-" * 30)

print("Вектор видимости:")
print("Формат:", visibility.shape)
print("Первые 5 элементов:", visibility[:5])
print("-" * 30)

print("Вектор количества осадков:")
print("Формат:", rainfall.shape)
print("Первые 5 элементов:", rainfall[:5])
rainfall_nonzero = rainfall[rainfall > 0]
first_5_nonzero_rainfall = rainfall_nonzero[:5]
print("Первые 5 ненулевых элементов:", first_5_nonzero_rainfall)
print("-" * 30)

print("Вектор количества снега:")
print("Формат:", snowfall.shape)
print("Первые 5 элементов:", snowfall[:5])
snowfall_nonzero = snowfall[snowfall > 0]
first_5_nonzero_snowfall = snowfall_nonzero[:5]
print("Первые 5 ненулевых элементов:", first_5_nonzero_snowfall)
print("-" * 30)

print("Вектор дат:")
print("Формат:", date_vector.shape)
print("Первые 5 элементов:", date_vector[:5])
print("-" * 30)

print("Матрица признаков :")
print("Формат:", feature_matrix.shape)
print("Первые 5 строк:\n", feature_matrix[:5])
print("-" * 30)

print("Матрица погодных условий: ")
print("Формат:", weather_matrix.shape)

non_zero_indices = np.where(np.all(weather_matrix != 0, axis=1))[0]

print("Первые 5 строк, где шел дождь снег и была солнечная радиация")
if len(non_zero_indices) > 0:
    for i in range(min(5, len(non_zero_indices))):
        index = non_zero_indices[i]
        print(f"Строка {index}: {weather_matrix[index]}")
else:
    print("Нет строк, где шел дождь снег и была солнечная радиация.")

print("-" * 30)

print("Матрица средних значений (для числовых столбцов):")
print("Формат:", mean_matrix_numeric.shape)
for i, col in enumerate(numeric_cols):
    print(f"Среднее значение для {col}: {mean_matrix_numeric[i]:.4f}")
print("-" * 30)

print("Матрица средних значений (для бинарных столбцов):")
print("Формат:", mean_matrix_binary.shape)
print("Средние значения:\n", mean_matrix_binary)
print("-" * 30)
```

##### 1.1.1 Коэффициент корреляции Пирсона

**Коэффициент корреляции Пирсона** — это статистическая мера, которая показывает степень линейной зависимости между двумя переменными. Он принимает значения от -1 до 1:

- **1** — означает полную положительную линейную корреляцию: при увеличении одной переменной другая переменная также увеличивается.
- **-1** — означает полную отрицательную линейную корреляцию: при увеличении одной переменной другая переменная уменьшается.
- **0** — означает отсутствие линейной корреляции между переменными.

Формула коэффициента корреляции Пирсона:

$$
\mathbf{r}_{XY} = \frac{\mathbf{cov}_{XY}}{\sigma_{X}\sigma_Y} = \frac{\sum(X-\bar{X})(Y-\bar{Y})}{\sqrt{\sum(X-\bar{X})^2\sum(Y-\bar{Y})^2}}.
$$

## Ковариация

**Ковариация** — это мера, которая показывает, насколько две случайные величины $X$ и $Y$ изменяются вместе. Она измеряет степень линейной зависимости между двумя переменными. Формула ковариации выглядит следующим образом:

$$
\mathbf{cov}_{XY} = \sum(X - \bar{X})(Y - \bar{Y}),
$$

где:
- $X$ и $Y$ — значения двух переменных,
- $\bar{X}$ и $\bar{Y}$ — средние значения переменных $X$ и $Y$ соответственно.

### Интерпретация ковариации:
- Если ковариация **положительная**, это означает, что при увеличении значений одной переменной значения другой переменной также имеют тенденцию увеличиваться.
- Если ковариация **отрицательная**, это означает, что при увеличении значений одной переменной значения другой переменной имеют тенденцию уменьшаться.
- Если ковариация **равна нулю**, это говорит об отсутствии линейной зависимости между переменными.


```python
def pearson_correlation(x, y):
    # Преобразуем входные данные в массивы NumPy
    x = np.array(x)
    y = np.array(y)

    # Проверяем, что векторы одинаковой длины
    if x.shape != y.shape:
        raise ValueError("Векторы должны быть одинаковой длины")

    # Вычисляем средние значения
    mean_x = np.mean(x)
    mean_y = np.mean(y)

    # Вычисляем ковариацию
    covariance = np.sum((x - mean_x) * (y - mean_y)) / len(x)

    # Вычисляем стандартные отклонения
    std_x = np.std(x)
    std_y = np.std(y)

    # Если одно из стандартных отклонений равно нулю, возвращаем 0
    if std_x == 0 or std_y == 0:
        return 0.0

    # Возвращаем коэффициент корреляции Пирсона
    return covariance / (std_x * std_y)
```

```python
pearson_temp_bike = pearson_correlation(temperature, bike_count)
print("-" * 30)
print("Корреляция Пирсона (температура и велосипеды):", pearson_temp_bike)

pearson_humidity_rainfall = pearson_correlation(humidity, rainfall)
print("Корреляция Пирсона (влажность и осадки):", pearson_humidity_rainfall)
print("-" * 30)
```

##### 1.1.2 Косинусное сходство
# Косинусное сходство

**Косинусное сходство** — это мера, используемая для определения степени сходства между двумя векторами в многомерном пространстве. Оно измеряет косинус угла между двумя векторами и часто применяется в анализе текстов, рекомендательных системах и других задачах машинного обучения.

Косинусное сходство принимает значения от $-1$ до $1$:
- **1** — означает, что векторы полностью совпадают по направлению (максимальное сходство).
- **0** — означает, что векторы ортогональны (отсутствие сходства).
- **-1** — означает, что векторы направлены в противоположные стороны (максимальное различие).

Формула косинусного сходства:

$$
\text{cosine_similarity}(A, B) = \frac{A \cdot B}{\|A\| \cdot \|B\|},
$$

где:
- $A \cdot B$ — скалярное произведение векторов $A$ и $B$,
- $\|A\|$ и $\|B\|$ — длины (нормы) векторов $A$ и $B$ соответственно.

## Скалярное произведение

**Скалярное произведение** двух векторов $A$ и $B$ вычисляется как сумма произведений их соответствующих компонент:

$$
A \cdot B = \sum_{i=1}^{n} A_i \cdot B_i,
$$

где $A_i$ и $B_i$ — компоненты векторов $A$ и $B$.

### Интерпретация косинусного сходства:
- Если косинусное сходство близко к **1**, векторы имеют одинаковое направление и высокую степень сходства.
- Если косинусное сходство близко к **0**, векторы ортогональны, и сходство отсутствует.
- Если косинусное сходство близко к **-1**, векторы направлены в противоположные стороны.


```python
def cosine_similarity(x, y):
    x = np.array(x).flatten()
    y = np.array(y).flatten()

    if x.shape != y.shape:
        raise ValueError("Векторы должны быть одинаковой длины")

    # Вычисляем скалярное произведение
    dot_product = np.dot(x, y)

    # Вычисляем нормы векторов
    norm_x = np.sqrt(np.sum(x**2))
    norm_y = np.sqrt(np.sum(y**2))

    # Если один из векторов нулевой, сходство не определено
    if norm_x == 0 or norm_y == 0:
        return 0.0

    return dot_product / (norm_x * norm_y)
```

```python
cosine_temp_bike = cosine_similarity(temperature, bike_count)

print("-" * 30)
print("Косинусное сходство (температура и велосипеды):", cosine_temp_bike)
print("-" * 30)

cosine_humidity_rainfall = cosine_similarity(humidity, rainfall)

print("Косинусное сходство (влажность и осадки):", cosine_humidity_rainfall)
print("-" * 30)
```

### **2. Нормализация векторов:**
# Нормализация векторов

**Нормализация вектора** — это процесс приведения вектора к единичной длине (норме), при котором направление вектора сохраняется, а его длина становится равной $1$. Нормализация часто используется в машинном обучении, анализе данных и других областях для упрощения вычислений и сравнения векторов.

## Формула нормализации вектора

Для нормализации вектора $A$ его компоненты делятся на его норму (длину). Формула нормализации вектора $A$ выглядит следующим образом:

$$
\hat{A} = \frac{A}{\|A\|},
$$

где:
- $\hat{A}$ — нормализованный вектор,
- $A$ — исходный вектор,
- $\|A\|$ — норма (длина) вектора $A$.

## Норма вектора

**Норма вектора** (длина вектора) вычисляется как квадратный корень из суммы квадратов его компонент:

$$
\|A\| = \sqrt{\sum_{i=1}^{n} A_i^2},
$$

где $A_i$ — компоненты вектора $A$.

### Пример нормализации

Пусть дан вектор $A = [3, 4]$. Его норма вычисляется как:

$$
\|A\| = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5.
$$

Тогда нормализованный вектор $\hat{A}$ будет:

$$
\hat{A} = \frac{A}{\|A\|} = \left[ \frac{3}{5}, \frac{4}{5} \right] = [0.6, 0.8].
$$



#### **2.1. Функция нормализации:**
- Напишите функцию для нормализации векторов данных до единичной
длины. Это может быть полезно для сравнения различных переменных,
таких как количество прокачанных велосипедов и погодные условия.


```python
def normalize_vector(vector):
    vector = np.array(vector, dtype=float)

    # Вычисляем норму вектора
    norm = np.sqrt(np.sum(vector**2))  # Норма = квадратный корень из суммы квадратов элементов

    # Если норма равна 0, возвращаем исходный вектор (избегаем деления на 0)
    if norm == 0:
        return vector

    # Нормализуем вектор
    return vector / norm
```

### **1. Фильтрация временных рядов:**

#### **1.1. Реализация фильтра:**
- Реализуйте простой фильтр временных рядов, используя скалярное
произведение. Примените его для сглаживания данных о прокате
велосипедов.

```python
def apply_kernel_filter(data, kernel):
    n = len(kernel)
    padded_data = np.pad(data, (n//2, n//2), mode='edge')
    filtered_data = np.array([np.dot(padded_data[i:i+n], kernel) for i in range(len(data))])
    return filtered_data
```


#### **1.2. Эксперименты с ядрами:**
- Экспериментируйте с разными ядрами (например, для выделения
трендов или сглаживания).

#### **1.2.1.  Прямоугольное ядро (Boxcar Kernel)**

Это ядро равномерно взвешивает все точки в окне. Оно подходит для простого усреднения.

#### Формула:
$$
K_{\text{boxcar}}(x) = \begin{cases}
\frac{1}{n}, & \text{если } |x| \leq \frac{n}{2} \\
0, & \text{иначе}
\end{cases}
$$

#### Пример (для $n = 5$):
$$
K_{\text{boxcar}} = \left[\frac{1}{5}, \frac{1}{5}, \frac{1}{5}, \frac{1}{5}, \frac{1}{5}\right]
$$

```python
def boxcar_kernel(size):
    return np.ones(size) / size
```

#### **1.2.2. Гауссово ядро**

Это ядро использует нормальное распределение для взвешивания точек. Оно подходит для плавного сглаживания.

#### Формула:
$$
K_{\text{gaussian}}(x) = \frac{1}{\sqrt{2\pi}\sigma} \exp\left(-\frac{x^2}{2\sigma^2}\right)
$$

#### Пример (для $n = 5$ и $\sigma = 1.0$):
$$
K_{\text{gaussian}} = [0.0545, 0.2442, 0.4026, 0.2442, 0.0545]
$$

```python
def gaussian_kernel(size, sigma=1.0):
    x = np.linspace(-size // 2, size // 2, size)
    kernel = np.exp(-x**2 / (2 * sigma**2))
    kernel /= np.sum(kernel)
    return kernel
```

#### **1.2.3. Ядро для выделения трендов**
Это ядро используется для выделения изменений (например, трендов). Оно вычисляет разницу между соседними точками.

#### Формула:
$$
K_{\text{trend}}(x) = \begin{cases}
-1, & \text{если } x = -\frac{n}{2} \\
1, & \text{если } x = \frac{n}{2} \\
0, & \text{иначе}
\end{cases}
$$

#### Пример (для $n = 5$):
$$
K_{\text{trend}} = [-1, 0, 0, 0, 1]
$$

```python
def trend_kernel(size):
    if size % 2 == 0:
        raise ValueError("Размер ядра должен быть нечетным.")

    kernel = np.zeros(size)
    kernel[size // 2 - 1] = -1
    kernel[size // 2 + 1] = 1
    return kernel
```

```python
# Ядро для выделения трендов
trend_kernel_5 = trend_kernel(5)

# Прямоугольное ядро
boxcar_kernel_5 = boxcar_kernel(5)

# Гауссово ядро
gaussian_kernel_5 = gaussian_kernel(5, sigma=1.0)

# Применение фильтров
trend_data = apply_kernel_filter(bike_count, trend_kernel_5)
smoothed_boxcar = apply_kernel_filter(bike_count, boxcar_kernel_5)
smoothed_gaussian = apply_kernel_filter(bike_count, gaussian_kernel_5)

print("-" * 30)
print("Тренд (первые 10 значений):")
print(trend_data[:10])
print("-" * 30)
print("Сглаженные данные (прямоугольное ядро, первые 10 значений):")
print(smoothed_boxcar[:10])
print("-" * 30)
print("Сглаженные данные (гауссово ядро, первые 10 значений):")
print(smoothed_gaussian[:10])
print("-" * 30)
```

### **2. Кластеризация методом k-средних:**
Метод k-средних — это алгоритм кластеризации, который группирует данные в $k$ кластеров на основе их близости к центроидам.

#### Формулировка задачи:
Дано множество точек данных $X = \{x_1, x_2, \dots, x_n\}$, где каждая точка $x_i$ — это вектор в $d$-мерном пространстве. Цель — разбить $X$ на $k$ кластеров $C = \{C_1, C_2, \dots, C_k\}$, минимизируя суммарное квадратичное отклонение точек от центроидов кластеров.

#### Математическая формулировка:
Алгоритм минимизирует целевую функцию:
$$
J = \sum_{i=1}^{k} \sum_{x \in C_i} \|x - \mu_i\|^2,
$$
где:
- $\mu_i$ — центроид кластера $C_i$, вычисляемый как среднее всех точек в кластере:
$$
\mu_i = \frac{1}{|C_i|} \sum_{x \in C_i} x.
$$

#### Шаги алгоритма:
1. **Инициализация**: Выберите $k$ начальных центроидов (например, случайно).
2. **Назначение**: Каждую точку $x_i$ назначьте ближайшему центроиду:
$$
C_i = \{x : \|x - \mu_i\|^2 \leq \|x - \mu_j\|^2 \text{ для всех } j \}.
$$
3. **Обновление**: Пересчитайте центроиды:
$$
\mu_i = \frac{1}{|C_i|} \sum_{x \in C_i} x.
$$
4. **Повторение**: Повторяйте шаги 2 и 3, пока центроиды не стабилизируются.

#### **2.1. Реализация алгоритма:**
- Реализуйте алгоритм кластеризации методом k-средних с
использованием NumPy, чтобы группировать дни по схожести в прокате
велосипедов.


```python
pip install matplotlib # Установим для красивой визуализации
```

```python
import matplotlib.pyplot as plt
```

```python
def k_means(data, k, max_iterations=100):

    # Инициализация центроидов (случайно выбранные точки)
    centroids = data[np.random.choice(data.shape[0], k, replace=False)]

    for i in range(max_iterations):
        # Назначение точек ближайшим центроидам
        distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # Пересчет центроидов
        new_centroids = np.array([data[labels == i].mean(axis=0) for i in range(k)])

        # Проверка на сходимость
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

k = 3
labels, centroids = k_means(data, k)

dataframe['Cluster'] = labels
print("-" * 30)
print("Центроиды кластеров:")
print(centroids)
print("-" * 30)
print("Первые 10 строк датафрейма с метками кластеров:")
print("-" * 30)
print(dataframe[['Date', 'Rented Bike Count', 'Cluster']].head(10))
print("-" * 30)
plt.scatter(data, np.zeros_like(bike_count), c=labels, cmap='viridis', s=50)
plt.scatter(centroids, np.zeros_like(centroids), c='red', marker='x', s=200, label='Центроиды')
plt.xlabel('Количество прокатов велосипедов')
plt.yticks([])
plt.legend()
plt.title(f'Кластеризация с k = {k}')
plt.show()
```

#### **2.2. Исследование влияния k:**
- Начните с фиксированного числа кластеров, а затем исследуйте, как влияет изменение значения k на результаты.

```python
k_values = range(1, 10)  # Значения k от 1 до 9
results = {}  # Для хранения результатов кластеризации

for k in k_values:
    labels, centroids = k_means(data, k)
    results[k] = {'labels': labels, 'centroids': centroids}

plt.figure(figsize=(20, 15))
for i, k in enumerate(k_values):
    plt.subplot(3, 3, i + 1)
    labels = results[k]['labels']
    centroids = results[k]['centroids']

    for cluster in range(k):
        cluster_data = data[labels == cluster]
        plt.scatter(cluster_data, np.zeros_like(cluster_data), label=f'Кластер {cluster + 1}')

    plt.scatter(centroids, np.zeros_like(centroids), c='red', marker='x', s=100, label='Центроиды')

    plt.title(f'k = {k}')
    plt.xlabel('Количество прокатов велосипедов')
    plt.yticks([])
    plt.grid(True)
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
df = dataframe

# 1. Зависимость проката велосипедов от температуры
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature(°C)'], df['Rented Bike Count'], alpha=0.5, color='blue')
plt.title('Зависимость количества арендованных велосипедов от температуры')
plt.xlabel('Температура (°C)')
plt.ylabel('Количество арендованных велосипедов')
plt.grid(True)
plt.show()

# 2. Прокат велосипедов по месяцам
df['Month'] = df['Date'].dt.month_name()

monthly_rentals = df.groupby('Month')['Rented Bike Count'].mean().reset_index()

months_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
monthly_rentals['Month'] = pd.Categorical(monthly_rentals['Month'], categories=months_order, ordered=True)
monthly_rentals = monthly_rentals.sort_values('Month')

plt.figure(figsize=(12, 6))
plt.bar(monthly_rentals['Month'], monthly_rentals['Rented Bike Count'], color='skyblue')
plt.title('Количество арендованных велосипедов по месяцам (среднее) ')
plt.xlabel('Месяц')
plt.ylabel('Количество арендованных велосипедов')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

monthly_rentals_total = df.groupby('Month')['Rented Bike Count'].sum().reset_index()
monthly_rentals_total['Month'] = pd.Categorical(monthly_rentals_total['Month'], categories=months_order, ordered=True)
monthly_rentals_total = monthly_rentals_total.sort_values('Month')

plt.figure(figsize=(12, 6))
plt.bar(monthly_rentals_total['Month'], monthly_rentals_total['Rented Bike Count'], color='lightgreen')
plt.title('Количество арендованных велосипедов по месяцам (суммарное)')
plt.xlabel('Месяц')
plt.ylabel('Количество арендованных велосипедов')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# 3. Прокат велосипедов по сезонам
seasons_map = {
    'Seasons_Spring': 'Spring',
    'Seasons_Summer': 'Summer',
    'Seasons_Autumn': 'Autumn',
    'Seasons_Winter': 'Winter'
}

df['Season'] = df[['Seasons_Spring', 'Seasons_Summer', 'Seasons_Autumn', 'Seasons_Winter']].idxmax(axis=1)
df['Season'] = df['Season'].replace(seasons_map)

seasonal_rentals = df.groupby('Season')['Rented Bike Count'].mean().reindex(['Spring', 'Summer', 'Autumn', 'Winter'])

plt.figure(figsize=(10, 6))
seasonal_rentals.plot(kind='bar', color=['#2ca02c', '#ff7f0e', '#d62728', '#1f77b4'])
plt.title('Количество арендованных велосипедов по сезонам (среднее)')
plt.xlabel('Сезон')
plt.ylabel('Количество арендованных велосипедов')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

seasonal_rentals_total = df.groupby('Season')['Rented Bike Count'].sum().reindex(['Spring', 'Summer', 'Autumn', 'Winter'])

plt.figure(figsize=(10, 6))
seasonal_rentals_total.plot(kind='bar', color=['#2ca02c', '#ff7f0e', '#d62728', '#1f77b4'])
plt.title('Количество арендованных велосипедов по сезонам (суммарное)')
plt.xlabel('Сезон')
plt.ylabel('Количество арендованных велосипедов')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# 4. Влажность и количество осадков
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.plot(df['Date'], df['Humidity(%)'], color='blue', label='Влажность (%)')
ax1.set_xlabel('Дата')
ax1.set_ylabel('Влажность (%)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.bar(df['Date'], df['Rainfall(mm)'] + df['Snowfall (cm)'],
        color='gray', alpha=0.3, label='Осадки (мм + см)')
ax2.set_ylabel('Осадки (мм + см)', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

plt.title('Влажность и количество осадков')
fig.legend(loc='upper left')
plt.show()

# 5. Прокат в зависимости от выходных дней
df['Holiday'] = np.where(df['Holiday_Holiday'] == 1, 'Выходные', 'Будни')

holiday_rentals = df.groupby('Holiday')['Rented Bike Count'].mean()

plt.figure(figsize=(8, 6))
holiday_rentals.plot(kind='bar', color=['#1f77b4', '#ff7f0e'])
plt.title('Среднее количество арендованных велосипедов: Выходные vs Будни')
plt.xlabel('Тип дня')
plt.ylabel('Среднее количество арендованных велосипедов')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

# 6. Скользящее среднее для временного ряда
df['Moving Average'] = df['Rented Bike Count'].rolling(window=7).mean()

plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Rented Bike Count'], label='Количество арендованных велосипедов', alpha=0.5)
plt.plot(df['Date'], df['Moving Average'], label='Скользящее среднее (7 дней)', color='red')
plt.title('Тренд количества арендованных велосипедов и скользящее среднее')
plt.xlabel('Дата')
plt.ylabel('Количество арендованных велосипедов')
plt.legend()
plt.grid(True)
plt.show()

# 7. Временной ряд (время - кол-во велосипедов)
time_series_bike_counts = np.array(df[['DateTime', 'Rented Bike Count']])
df['DateTime'] = pd.to_datetime(df['DateTime'])

plt.figure(figsize=(14, 7))
plt.plot(df['DateTime'], df['Rented Bike Count'], label='Количество арендованных велосипедов', color='blue', alpha=0.7)
plt.title('Временной ряд количества арендованных велосипедов')
plt.xlabel('Дата и время')
plt.ylabel('Количество арендованных велосипедов')
plt.legend()
plt.grid(True)
plt.show()
```
